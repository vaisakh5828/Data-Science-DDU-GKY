# One hot encoding and Feature Scaling

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Loding the dataset

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data
```

### One Hot Encoding

```python
data['SEX'] = pd.get_dummies(data=data.sex, drop_first=True)
```

```python
data.head(3)
```

### Converting the Entire dataframe

```python
data_No_Dummies = pd.get_dummies(data=data)
```

```python
data_No_Dummies
```

```python
data_No_Dummies.shape
```

```python
data_With_Dummies = pd.get_dummies(data=data, drop_first=True)
data_With_Dummies.head()
```

```python
data_With_Dummies.shape
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, LabelEncoder
```

## Data Scaling or Data Normalising

## Minmax Scaler

```python
minmax = MinMaxScaler()
```

```python
data[['age']]
```

```python
data[['age']].values
```

```python
age = data[['age']].values
```

```python
age
```

```python
data['AGE'] = minmax.fit_transform(age)
```

```python
data.head()
```

```python
data.AGE.min()
```

```python
data.AGE.max()
```

## Standard Scaler

```python
sc = StandardScaler()
```

```python
data['AGE_SC'] = sc.fit_transform(age)
```

```python
data
```

```python
data.AGE_SC.min()
```

```python
data.AGE_SC.max()
```
