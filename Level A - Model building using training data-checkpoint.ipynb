### Problem statement :
It is a simple dataset that can be used to determine the likelihood that a candidate that has
been trained for a company will eventually leave the company or stick with
them.


<li> Feature explanation is below:
● enrollee_id : Unique ID for candidate
● city: City code
● city_ development _index : Developement index of the city (scaled)
● gender: Gender of candidate
● relevent_experience: Relevant experience of candidate
● enrolled_university: Type of University course enrolled if any
● education_level: Education level of candidate
● major_discipline :Education major discipline of candidate
● experience: Candidate total experience in years
● company_size: No of employees in current employer's company
<li> company_type : Type of current employer
● lastnewjob: Difference in years between previous job and current job
● training_hours: training hours completed
● target: 0 – Not looking for job change, 1 – Looking for a job change

#### Importing necessary libraries

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from imblearn.over_sampling import RandomOverSampler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, auc, confusion_matrix, roc_auc_score, recall_score,classification_report,plot_roc_curve
import pickle
```

#### Loading the training data set dataset

```python
data = pd.read_csv('./aug_train.csv')
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.columns
```

#### Checking the null percentage

```python
data.isnull().mean()*100
```

#### Dropping the null values

```python
data.dropna(inplace=True)
data.head(10)
```

#### Dropping the unwanted column 'enrollee_id'

```python
data.drop(columns='enrollee_id', inplace=True)
```

```python
data.head(10)
```

#### Encoding the categorical columns

```python
LE = LabelEncoder()
```

```python
le_columns = ['city', 'gender', 'relevent_experience', 'enrolled_university', 'education_level', 'major_discipline', 'company_type']
```

```python
for i in le_columns:
    data[i] = LE.fit_transform(data[i])
```

```python
data.head()
```

#### Treating the experience, company_size, last_new_job columns 

```python
data.company_size.unique()
```

```python
data.company_size.replace({'50-99':99, '<10':9, '5000-9999':9999, '1000-4999':4999, '10/49':49, '100-500':500,
       '10000+':10000, '500-999':1000}, inplace=True)
```

```python
data.experience.unique()
```

```python
data.experience.replace({'15':15, '>20':21, '13':13, '7':7, '5':5, '16':16, '11':11, '<1':0, '18':18, '19':19, '12':12,
       '10':10, '9':9, '2':2, '6':6, '4':4, '14':14, '3':3, '8':8, '17':17, '20':20, '1':1}, inplace=True)
```

```python
data.last_new_job.unique()
```

```python
data.last_new_job.replace({'>4':5, '4':4, '1':1, '3':3, '2':2, 'never':0}, inplace=True)
```

```python
data.info()
```

#### Changing the data type of target column into integer

```python
data['target'] = data.target.astype(int)
```

```python
data.head(10)
```

#### Selecting features and targets

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

#### Checking the balance of target column

```python
sns.countplot(data=data, x= 'target')
plt.show()
```

<li> We can see that the dataset is unbalanced. We may treat it using over sampling method.

```python
oversample = RandomOverSampler(sampling_strategy='minority')
```

```python
x_over, y_over = oversample.fit_resample(x, y)
```

```python
sns.countplot(x = y_over)
plt.show()
```

#### Feature scaling

```python
SC = StandardScaler()
```

```python
x = SC.fit_transform(x_over)
y = y_over
```

#### Splitting the data into training and testing sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)
```

### Machine learning

#### Model building

```python
models = {
    "LogisticRegression" : LogisticRegression(),
    "RandomForestClassifier" : RandomForestClassifier(),
    "KNeighborsClassifier"   : KNeighborsClassifier(),
    "SVC" : SVC()
}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    print("model trained with {}".format(name))
    score_train =accuracy_score (y_train, model.predict(x_train))
    score_tst = accuracy_score(y_test, model.predict(x_test))
    con_train = confusion_matrix(y_train, model.predict(x_train))
    con_test = confusion_matrix(y_test, model.predict(x_test))
    print("Model accuracy on train is:: ",score_train)
    print("Model accuracy on test is:: ", score_tst)
    print("confusion_matrix train is:: ", con_train)
    print("confusion_matrix test is:: ", con_test)
    print("Wrong Predictions made:",(y_test !=model.predict(x_test)).sum(),'/',((y_test == model.predict(x_test)).sum()+(y_test != model.predict(x_test)).sum()))
    print("*"*60)
```

#### We may deduce that RandomForestClassifier has the best perfomance.

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
print(classification_report(y_test, y_predict))
```

### Saving the actual and predicted values to a dataframe

```python
PREDICTIONS =pd.DataFrame(zip(y_test, y_predict), columns=["Actual", "Predicted"])
PREDICTIONS
```

### ROC Curve

```python
plot_roc_curve(model, x_test, y_test);
```
