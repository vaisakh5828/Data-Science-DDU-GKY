# Plotting with pandas

```python
import pandas as pd 
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./Car_sales.csv', parse_dates=['Latest_Launch'])
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.sort_values(by='Latest_Launch', inplace=True)
```

```python
data
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head(10)
```

### Extracting the dates 

```python
data["Year"] = data.Latest_Launch.dt.year
data['Month'] = data.Latest_Launch.dt.month
data['Day'] = data.Latest_Launch.dt.day
```

```python
data.head()
```

## Univariate Analysis

### Histogram

```python
data.Sales_in_thousands.plot.hist(cmap = 'Reds');
```

## Box plot

```python
data.Sales_in_thousands.plot.box(cmap ='magma');
```

## Kernel Density Estimation Plot (KDE PLOT)

```python
data.Sales_in_thousands.plot.kde(cmap = 'flag');
```

### Sub plots 

```python
data.plot.box( figsize=(18,3), subplots = True);
```

```python
data.plot.hist( figsize=(6,12), subplots = True);
```

```python
data.plot.kde(figsize = (5,15),subplots = True);
```

## Line charts

```python
data.plot.line(x='__year_resale_value', y='Price_in_thousands', cmap = 'Spectral');
```

#### This happens due to the fact that year is not in order.

```python
data.sort_values(by='__year_resale_value').plot.line(x='__year_resale_value', y='Price_in_thousands', cmap='Wistia');
```

```python
data.plot.bar(x='Manufacturer', y='Sales_in_thousands', figsize = (20, 5),cmap = 'gist_earth');
```

### Count plot

```python
data.Manufacturer.value_counts().plot.bar(figsize=(12,5), cmap = 'Paired');
```

## Area chart

```python
data.plot.area(x='Manufacturer', y='Sales_in_thousands', figsize = (10,6), cmap = 'rainbow');
```

## Horizontal bar chart

```python
data.sort_values(by='Price_in_thousands')[:5].plot.barh(x='Model', y= 'Power_perf_factor', figsize = (6,4), cmap = 'PuBu');
```

## Scatter plot

```python
data.sort_values(by='Price_in_thousands')[:5].plot.scatter(x='Latest_Launch', y='Price_in_thousands', figsize = (6,6));
```

### Pie chart

```python
data.Vehicle_type.value_counts().plot.pie();
```

```python
data.Vehicle_type.value_counts().plot.pie(autopct ='%1.1f%%', legend = True);
```

## Hexbin plot

```python
data.plot.hexbin(x='Sales_in_thousands', y='Engine_size');
```

```python
data.plot.hexbin('Curb_weight', 'Fuel_capacity', gridsize=25);
```
