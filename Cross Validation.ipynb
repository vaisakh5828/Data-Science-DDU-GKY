### Setting the outlier limits

```python
def outlier_limit(col):
    Q3,Q1 = np.nanpercentile(col, [75,25])
    IQR = Q3-Q1
    UL = Q3+1.5*IQR
    LL = Q1-1.5*IQR
    return UL,LL
```

```python
for i in data.columns:
    if data[i].dtypes != "object":
        UL,LL = outlier_limit(data[i])
        data[i] = np.where((data[i]>UL)|(data[i]<LL), np.nan, data[i])
```

```python
for i in data.columns:
    if data[i].dtypes != "object":
        sns.boxplot(y = data[i], color="green")
        plt.show()
```

#### Lets take only the columns with good correlation with the target

```python
x= data[[ 'hdlngth', 'skullw', 'totlngth','earconch', 'chest', 'belly']].values
```

```python
y = data[['footlgth']].values
```

### Feature Scaling 

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

### Splitting the data into training and testing set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=55)
```

### Model building

```python
LR = LinearRegression()
```

```python
LR.fit(x_train, y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

### Cross validation

```python
scores = cross_val_score(LR, x,y, scoring="neg_mean_squared_error",cv=10,n_jobs=-1)
```

```python
scores
```

```python
rmse = np.sqrt(-scores)
```
