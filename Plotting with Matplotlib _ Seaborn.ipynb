# <center> Plotting with Matplotlib & Seaborn

```python
import pandas as pd 
import numpy as np 
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./Car_sales.csv')
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

## Univariate Analysis

### Histogram

```python
sns.histplot(data=data, x='Sales_in_thousands')
plt.show()
```

## Box plot

```python
sns.boxplot(data=data, x='Sales_in_thousands')
plt.show()
```

```python
sns.boxplot(data=data, y='Sales_in_thousands')
plt.show()
```

## Kernel Density Estimation Plot (KDE PLOT)

```python
sns.kdeplot(data=data.Sales_in_thousands)
plt.show()
```

### Sub plots 

```python
sns.histplot(data=data)
plt.show()
```

```python
plt.figure(figsize=(18,6))
sns.boxplot(data=data)
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.xticks(rotation = 90)
sns.boxplot(data=data)
plt.show()
```

```python
sns.kdeplot(data=data)
plt.show()
```

### For loop to plot subplots

```python
for i in data.columns:
    if data[i].dtypes != 'object':
        sns.boxplot(y=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtypes != 'object':
        sns.histplot(x=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtypes != 'object':
        sns.kdeplot(data=data[i])
        plt.show()
```

### Subplots in matplotlib

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=data,y='Sales_in_thousands')

plt.subplot(2,2,2)
sns.boxplot(data=data,y='Price_in_thousands')

plt.subplot(2,2,3)
sns.boxplot(data=data,y='Engine_size')

plt.subplot(2,2,4)
sns.boxplot(data=data,y='Horsepower')

plt.show()
```

### Subplots using enumerate 

#### Let's create list in which there is only the columns which we are going to plot

```python
features = []
```

```python
for i in data.columns:
    if data[i].dtypes != 'object':
        features.append(i)
```

```python
features
```

```python
list(enumerate(features))
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(4,3,i[0]+1)
    sns.boxplot(data=data, y=i[1], color='red')
```

## Line charts

```python
plt.figure(figsize=(10,6))
plt.xlabel('Year Resale Value')
plt.ylabel('Price')
plt.title('GRAPH')
sns.lineplot(data=data,x='__year_resale_value', y='Price_in_thousands')
plt.show()
```

### Count plot

```python
plt.figure(figsize=(16,5))
plt.xticks(rotation = 90)
sns.countplot(data=data,x='Manufacturer')
plt.show()
```

## Area chart

```python
plt.figure(figsize=(18,5))
plt.xticks(rotation = 90)
plt.stackplot(data.Manufacturer,data.Sales_in_thousands)
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.xticks(rotation = 90)
plt.stackplot(data.Manufacturer,data.Sales_in_thousands, data.Price_in_thousands)
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.xticks(rotation = 90)
plt.stackplot(data.Manufacturer,data.Sales_in_thousands, data.Price_in_thousands, data.Horsepower)
plt.show()
```

## Horizontal bar chart

```python
plt.figure(figsize=(6,12))
plt.xticks(rotation = 90)
sns.barplot(data=data, x='Sales_in_thousands', y='Manufacturer')
plt.show()
```

## Vertical Bar plot

```python
plt.figure(figsize=(12,4))
plt.xticks(rotation = 90)
sns.barplot(data=data, x='Manufacturer', y='Horsepower')
plt.show()
```

## Scatter plot

```python
sns.scatterplot(data=data, x='Horsepower', y='Sales_in_thousands')
plt.show()
```

```python
sns.scatterplot(data=data, x='Horsepower', y='Sales_in_thousands', hue='Vehicle_type')
plt.show()
```

### Pie chart

```python
data.Vehicle_type.value_counts().index
```

```python
palette_color = sns.color_palette('dark')
```

```python
plt.pie(data=data, x=data.Vehicle_type.value_counts(),explode = [0.1, 0], colors=palette_color,labels=['Passenger', 'Car'],autopct='%0.2f%%');
```

### Use ' \ ' for long codes to write in lines seperate

```python
plt.pie(data=data,\
        x=data.Vehicle_type.value_counts(),\
        explode = [0.1, 0], \
        colors=palette_color,\
        labels=['Passenger', 'Car'],\
        autopct='%0.2f%%');
```

## Correlation matrix or Heatmap

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

#### More cooler approach

```python
sns.heatmap(data.corr()[["Price_in_thousands"]].sort_values(by="Price_in_thousands", ascending=False), vmin=-1, vmax=1,annot=True, cmap="Greens")
plt.plot();
```
