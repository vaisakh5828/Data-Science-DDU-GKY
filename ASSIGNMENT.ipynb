```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('./Final.csv',encoding='latin')
```

```python
df
```

#### Data Cleaning

```python
df.isnull().sum()
```

```python
df.Rating.fillna(df.Rating.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Average']=df.Average.astype('str')
df['Average']=df.Average.str.replace('\D',"",regex=True)
df['Average']=df.Average.astype('int')
```

```python
slry_yr=df.query("Category=='year'")
```

```python
slry_mnth=df.query("Category=='month'")
```

```python
slry_hr=df.query("Category=='hour'")
```

#### Q1.Which Top Rated companys are providing more average salary(yearly)

```python
high_rating=df.query("Rating > 4.0 ")
```

### Q2 What is the average salary for data scientists in the dataset? Are there any significant variations in salaries across different categories or companies?

#### Distribution of Rating

```python
average_salary = df['Average'].mean()
print("Average Salary for Data Scientists:", average_salary)
```

```python
average_salary_by_category = df.groupby('Category')['Average'].mean()
print(average_salary_by_category)
```

```python
average_salary_by_company = df.groupby('Company')['Average'].mean()
print(average_salary_by_company)
```
