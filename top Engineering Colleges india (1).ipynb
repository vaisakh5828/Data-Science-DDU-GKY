### Analysis of the Indian college and ranking 2022

### India is 2nd world's largest population country and every year lacks student get their college degree from various institutes in India I scrap a website containing the top 200 colleges in India in 2022, a dataset containing a list of colleges from rank 1 to 200 and some other features,


##### the feature is explained below:-
1)rank:- 1-200 ranking of the college according to different parameter

2)name:- the name of the respective college

3)ownership:- which organization own the college there are two values either public or private

4)grade:- the grade of college higher count of 'A' the good score will max(5)

5)TLR(Teaching-learning resources)= define resources as materials used by a teacher to supplement classroom instruction or to stimulate the interest of learners in simple quality content delivered to students.

6)RPC(Research and Professional Practice) developing knowledge and experience, which are standard within a field or profession, and involves information-gathering procedures that are a part of the typical relationship between practitioners

7)GO(graduation Outcome)= to understand whether recent graduates are in employment, have continued with further study or are doing something else

8)OI(outreach and inclusivity)=> it contains the following factors the number of students from another state or countries, per cent of women, facility for physically challenged student

9)perc(peer perception):- college perception in the market, stakeholder, facility perception, student perception, perception of other universities for you.

10)total:- overall percentage

### list of govt colleges

```python
govt_colleges=data.query("owner_ship =='Public/Government'" )
```

```python
data.owner_ship	.unique()
```

```python
data.grade.replace({'none':1,'AA':2,'AAA':3,'AAAA':4,'AAAAA':5},inplace=True)
```

```python
data['grade']=data.grade.astype('int')
```

```python
data['ownership_int']=pd.get_dummies(data=data.owner_ship,drop_first=True)
```

```python
plt.title("GRAPH NO: :rank allocation of Government Colleges")
sns.barplot(data=govt_colleges,x='grade',y='rank')
plt.show()
```

### What is the distribution of college grades in the dataset?

## conclusion

Based on the available data, an analysis of the top engineering colleges in India reveals the following:

Rank: The ranking column indicates the position of each college. By analyzing this column, we can identify the top-ranked engineering colleges in India.

Ownership: The ownership column provides information about the ownership types of the colleges, such as government-funded, private, or autonomous. This analysis helps understand the distribution of ownership types among the top-ranked colleges.

Grade: The grade column represents the quality or accreditation level of each college. Analyzing this column allows for an assessment of the overall quality and performance of the top engineering colleges.

TLR: The Teaching, Learning, and Resources (TLR) column provides insights into the educational aspects of the colleges. It includes factors such as teaching quality, learning environment, and available resources.

RPC: The Research, Professional Practice, and Collaborative Performance (RPC) column may contain data related to research output, industry collaborations, and professional practices of the colleges. Analyzing this column offers insights into the research and industry engagement of the top-ranked colleges.

GO: The GO column represents the Graduate Outcome or employability of the students from each college. Analyzing this column provides an idea of the job placements and career prospects for graduates of the top engineering colleges.

OI: The Overall Inclusivity (OI) column indicates the inclusiveness and diversity of the colleges. Analyzing this column offers insights into the colleges' efforts to create an inclusive environment for students from different backgrounds.

Perc: The Perc column may represent the percentage or score of various performance indicators. It is crucial to understand the specific indicators being measured to draw meaningful conclusions.

Total: The Total column likely represents an aggregate score or overall ranking based on the combined performance of each college. Analyzing this column can help identify the top-performing engineering colleges in India.
